```python
import pandas as pd
```

```python
dataset = pd.read_csv('DataSet/Train_aBjfeNk.csv')
```

```python
dataset.head(), dataset.shape
```

```python
dataset.Segmentation.unique()
```

```python
data = dataset.sample(frac=0.95, random_state=786)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.classification import *

exp_clf101 = setup(data = data, target = 'Segmentation', session_id=123)
```

```python
compare_models()
```

```python
models()
```

```python
xgboost = create_model('xgboost')
```

```python
tuned_xgb = tune_model(xgboost)
```

```python
predict_model(tuned_xgb)
```

```python
unseen_predictions = predict_model(tuned_xgb, data=data_unseen)
unseen_predictions.head()
```

```python
from pycaret.utils import check_metric
check_metric(unseen_predictions.Segmentation, unseen_predictions.Label, 'Accuracy')
```

```python
save_model(tuned_xgb,'Final XGB Model 01Sep2020')
```
